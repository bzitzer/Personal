```python
# Generate some random numbers, gaussian distributed, then calculate mean and std. dev.
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn

val_max = 10000
values = randn(val_max)
#print(values)
y,x = np.histogram(values,bins=np.arange(-6,6,0.025))
print(x)
#print(y)
fig, ax = plt.subplots()
ax.plot(x[:-1],y)
fig.show()
mean = 0
sum = 0
for i in range(1,val_max):
    mean += values[i]

mean = mean/val_max
std_dev = 0    
for i in range(1,val_max):
    std_dev += pow(mean - values[i],2)
    
integral = 0
for i in range(1,x.size-1):
    if x[i] < 1 and x[i] > -1:
        integral += y[i]

print('integral between +/- 2 sigma',integral/val_max)
std_dev = np.sqrt(std_dev/(val_max-1))
print(mean)
print(std_dev)
```

```python
import math as mt

N = 99 # Number of cards in deck
K = 34 # Number of lands in deck (sucesses)
n = 7 # Number of cards in hand
k = 1 # Number of desired sucesses in hand

K_Cr_k = mt.factorial(K)/(mt.factorial(k)*mt.factorial(K-k))
NK_Cr_nk = mt.factorial(N-K)/(mt.factorial(n-k)*mt.factorial(N-K - (n-k)))
N_Cr_n = mt.factorial(N)/(mt.factorial(n)*mt.factorial(N-n))

P_k = K_Cr_k*NK_Cr_nk/N_Cr_n
print('Probability of exactly',k,'successes in a',n,'card hand with',K,'lands:', P_k)
P_lt_k = 0
P_gt_k = 0
for i in range(0,n):
    if(i>K):
        break
    K_Cr_k = mt.factorial(K)/(mt.factorial(i)*mt.factorial(K-i))
    NK_Cr_nk = mt.factorial(N-K)/(mt.factorial(n-i)*mt.factorial(N-K - (n-i)))
    if i < k:
        P_lt_k += K_Cr_k*NK_Cr_nk/N_Cr_n
    else:
        P_gt_k += K_Cr_k*NK_Cr_nk/N_Cr_n

print('Probability of less than',k,'successes in a',n,'card hand with',K,'lands:', P_lt_k)
print('Probability of at least',k,'successes in a',n,'card hand with',K,'lands:', P_gt_k)
#print('sum',P_lt_k+P_gt_k)
```

```python
import math as mt
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.stats import norm
import numpy as np
import matplotlib.mlab as mlab

def gaus(x,a,x0,sigma):
    return a*exp(-(x-x0)**2/(2*sigma**2))

def lima(Non,Noff,alpha):
    if(Non == Noff*alpha):
        return(0)
    S = mt.sqrt(2)*mt.sqrt(Non*mt.log(Non*(1+alpha)/(alpha*(Non+Noff))) + Noff*mt.log(Noff*(1+alpha)/(Non+Noff)))
    if(Non < Noff*alpha):
        S *= -1.0
    return(S)

Rbg = 3.0 # bg rate 1/min
Exp = 100.0*60 # minutes 
alpha = 0.2

NMax = 10000
#print(np.random.poisson(int(Rbg*Exp),100))
Non = np.random.poisson(Rbg*Exp,NMax)
Noff = np.random.poisson(Rbg*Exp/alpha,NMax)
S_array = np.array([])
for i in range(1,NMax):
    #print('Non',Non[i],'Noff',Noff[i],'Excess',Non[i]-alpha*Noff[i])
    S = lima(Non[i],Noff[i],alpha)
    S_array = np.append(S_array,[S])
    #print("Significance",S)
(mu,sigma) = norm.fit(S_array)

print('mu',mu,'sigma',sigma)
print('Array',S_array)
n, bins, patches = plt.hist(S_array,normed=True,bins=100)
print('bins',bins)
print('n',n)
y = mlab.normpdf(bins,mu,sigma)
#l = plt.semilogy(bins, y, 'r--', linewidth=2)
l = plt.plot(bins,y,'r--',linewidth=2)
plt.show()
```
