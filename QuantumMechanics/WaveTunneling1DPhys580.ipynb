```python
import numpy as np
import math as mt
import cmath as cmt
import matplotlib.pyplot as plt

def init_waveform(x0,sigma,k0 = 500,n = 10000,x_min = -1,x_max = 1):
    # initial 1D wave packet
    x = np.linspace(x_min,x_max,n)
    y = np.zeros(n,dtype=complex)
    norm = 1/(mt.sqrt(2*mt.pi)*sigma)
    test_norm = 0
    j = cmt.sqrt(-1)
    dx = (x_max - x_min)/n
    for i in range(1,n):
        y[i] = cmt.exp(-0.5*(x[i] - x0)**2/sigma**2 - k0*j*(x[i]-x0))
        #y[i] *= norm
        test_norm += y[i]*dx
    print('norm: ',test_norm)
    return(y)

def V_x(x,V0 = 100,w = 0.1):
    # functional form of the potential barrier, height V0 and width w
    if x<-w/2:
        return(0)
    if x<w/2:
        return(V0)
    else:
        return(0)
    
# setting up constants
n_max = int(1.2/5e-3)
x_min = -0.6
x_max = 0.6 
x     = np.linspace(x_min,x_max,n_max)
x0    = -0.4
sigma = 0.05
y0 = init_waveform(x0,sigma,500,n_max,x_min,x_max)
plt.plot(x,np.real(y0)**2+np.imag(y0)**2)
plt.show()
print(V_x(-0.5),V_x(0.5,10))
```

```python

#np.seterr(all='ignore')
t_max = 0.02
#dt = 1e-5
m_max = int(2e5)
dt = t_max/m_max

dx = (x_max - x_min)/n_max
y = init_waveform(-0.3,0.05,50,n_max,x_min,x_max)
R = np.real(y)
I = np.imag(y)
norm = mt.sqrt(np.sum(R**2 + I**2)*dx)
R /= norm
I /= norm
print('norm at step 0: ',np.sum(R**2)*dx+np.sum(I**2)*dx)

dR = np.zeros(n_max)
dI = np.zeros(n_max)

plt.ion()
fig = plt.figure()
#plt.plot(x,P*dx)
numImage = 0

ax = plt.axes([-0.6,0.6,0.0,0.06])
plt.ylim(0,0.06)
for m in range(1,m_max+1): # loop over time steps
    P = R**2 + I**2
    norm = mt.sqrt(np.sum(P)*dx)
    
    if m%1000 == 0: # update user info
        str1 = "On timestep: " + str(m)
        print(str1)
        print('norm: ',np.sum(P)*dx)
        S = np.sum(P*np.log(P*dx)*dx)
        print('Entropy: ',S)
    if m%1000 == 0: # save plot, calc transmittance and reflection
        fileNm = 'qmWave1Dtimestep' + str(numImage) + '.png' 
        plt.close()
        plt.ylim(0,0.06)
        plt.xlabel('x position')
        plt.ylabel('Probability density')
        Tr = 100.0*np.sum(P[int(n_max/2)+1:n_max])*dx
        Re = 100.0*np.sum(P[0:int(n_max/2)])*dx
        plt.text(-0.5,0.05,'Transmitted: '+str(round(Tr,3))+' %')
        plt.text(-0.5,0.045,'Reflected: '+str(round(Re,3))+' %')
        
        plt.plot(x,P*dx)
        
        plt.savefig(fileNm)
        print('saved ',fileNm)
        numImage += 1
        
    for n in range(0,n_max-1): # loop over x positions
        # second derivative in x
        lapx_I = (I[n+1] + I[n-1] - 2*I[n])/dx**2
        lapx_R = (R[n+1] + R[n-1] - 2*R[n])/dx**2
        # leapfrog to update the real and imaginary parts using schrodinger's equation
        dR[n] =  dt*(0.5*lapx_I + V_x(x[n],1e4,0.05)*I[n]) 
        dI[n] = -dt*(0.5*lapx_R + V_x(x[n],1e4,0.05)*R[n])
    for n in range(0,n_max-1): 
        # update for next time step
        R[n] += dR[n]
        I[n] += dI[n] 
        R[n] /= norm
        I[n] /= norm
   


#plt.plot(x,P*dx)
plt.show()
        
print(np.sum(P)*dx)
print('Transmission: ',np.sum(P[int(n_max/2)+1:n_max])*dx)
print('Reflection: ',np.sum(P[0:int(n_max/2)])*dx)
```
